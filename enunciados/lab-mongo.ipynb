```python
# importamos las librerías que necesitamos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd
import numpy as np

# Para trabajar con datos de tipo fecha
# ------------------------------------------------------------------------------
from datetime import datetime

# Para trabajar con Mongo
# ------------------------------------------------------------------------------
import pymongo

# Para utilizar nuestras funciones
# ------------------------------------------------------------------------------
#from src import soporte_mongo as sm

# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
# lo primero que tenemos que hacer es conectarnos con Mongo
cliente = pymongo.MongoClient()
print(cliente)

# como la conexión se ha realizado con éxito vamos a ver las bases de datos que tenemos
cliente.list_database_names()
```

```python
# nos conectamos con una BBDD 
EcommerceDB = cliente['Ecommerce']

#Veamos además las colecciones de la BBDD, usando el método 'list_collection_names()'
EcommerceDB.list_collection_names()
```

```python
Companies = EcommerceDB['Companies']
```

```python
resultado_companies = pd.DataFrame(Companies.find())
display(resultado_companies.head(1))
resultado_companies.shape[0]
```

### 1. Todas las empresas cuyo nombre sea 'Babelgum'. Recupera solo su campo `name`.


```python
find_babelgum = pd.DataFrame(Companies.find({'name': {"$eq": 'Babelgum'}},{'name': 1}))
find_babelgum
```

### 2. Todas las empresas que tienen más de 5000 empleados. Limita la búsqueda a 20 empresas y ordénalas por **número de empleados**.


```python
find_emp_5000 = pd.DataFrame(Companies.find({"number_of_employees" : {"$gt": 5000}}).sort({'number_of_employees':1}).limit(20))
find_emp_5000
```

### 3. Todas las empresas fundadas entre 2000 y 2005, incluyendo ambos años. Recupera solo los campos `name` y `founded_year`.


```python
# solo name y founded_year
#empresas entre 2005 y 2000
mayor_2000 = {"founded_year": {"$gte": 2000.0}}
menor_2005 = {"founded_year": {"$lte": 2005.0}}
fund_2000_2005 = pd.DataFrame(Companies.find({'$and':  [mayor_2000, menor_2005]}, {'name': 1, 'founded_year': 1})) 
fund_2000_2005
```

### 4. Todas las empresas que tuvieron una valoración de más de 100.000.000 y que fueron fundadas antes de 2010. Recupera solo los campos `name` e `ipo`.


```python
mayor_valor = {"ipo.valuation_amount": {"$gt": 100000000}}
menor_2010 = {"founded_year": {"$lt": 2010.0}}
fund_100_2010 = pd.DataFrame(Companies.find({'$and': [mayor_valor, menor_2010]}, {'name': 1, 'ipo': 1, '_id':0})) 
fund_100_2010
```

### 5. Todas las empresas que tienen menos de 1000 empleados y que fueron fundadas antes de 2005. Ordénalas por el número de empleados y limita la búsqueda a 10 empresas.

```python
emp_menos_1000 = {"number_of_employees": {"$lt": 1000}}
antes_2005 = {"founded_year": {"$lt": 2005.0}}
empl_antes_2005 = pd.DataFrame(Companies.find({'$and': [emp_menos_1000, antes_2005]}).sort({'number_of_employees':-1}).limit(10))
empl_antes_2005
```

### 6. Todas las empresas que no incluyen el campo `partners`.

```python
partners_out = pd.DataFrame(Companies.find({'partners': {"$exists": 'False'}}))
# find_exists_df = pd.DataFrame(bcn.find({'stars': {'$exists': False}}))
partners_out
```

### 7. Todas las empresas que tienen un valor nulo en el campo `category_code`.

```python
cod_categoria = pd.DataFrame(Companies.find({'category_code': None}))
cod_categoria
```

### 8. Todas las empresas que tienen al menos 100 empleados pero menos de 1000. Recupera solo los campos `name` y `number of employees`.

```python

entre_100_1000 = pd.DataFrame(Companies.find({'number_of_employees': {"$gte": 100, "$lt": 1000}},{'name': 1, 'number_of_employees': 1}))
entre_100_1000
```

### 9. Ordena todas las empresas por su precio de IPO en orden descendente.

```python
ipo_desc = pd.DataFrame(Companies.find({"ipo.valuation_amount": {"$exists": 'True'}}).sort({"ipo.valuation_amount":-1}))
ipo_desc
```

### 10. Recupera las 10 empresas con más empleados, ordenadas por el `número de empleados`.

```python
empresas_10_top = pd.DataFrame(Companies.find().sort({'number_of_employees':-1}).limit(10))
empresas_10_top
```

### 11. Todas las empresas fundadas en el segundo semestre del año. Limita tu búsqueda a 1000 empresas.

```python
segundo_semestre = pd.DataFrame(Companies.find({'founded_month': {"$gt": 6}}).limit(1000))
segundo_semestre
```

### 12. Todas las empresas fundadas antes del año 2000 que tienen un monto de adquisición superior a 10.000.000.

```python
antes_2000 = {"founded_year": {"$lt": 2000.0}}
adquision_ = {"acquisition.price_amount": {"$gt": 10000000}}
antes_2000_adcq = pd.DataFrame(Companies.find({'$and': [antes_2000, adquision_]}))
antes_2000_adcq
```

### 13. Todas las empresas que han sido adquiridas después de 2010, ordénalas por el monto de adquisición, y recupera solo los campos `name` y `acquisition`.

```python
despues_2010 = pd.DataFrame(Companies.find({"acquisition.acquired_year" : {"$gt": 2010}}, {'name':1, 'acquisition': 1}).sort({'acquisition.price_amount':-1}))
despues_2010
```

### 14. Ordena las empresas por su `founded year`, recuperando solo los campos `name` y `founded year`.

```python
ano_fundado = pd.DataFrame(Companies.find([], {'name': 1, 'founded_year':1}).sort({'founded_year':1}))
ano_fundado
```

### 15. Todas las empresas que han sido fundadas en los primeros siete días del mes, incluyendo el séptimo. Ordénalas por su `acquisition price` en orden descendente. Limita la búsqueda a 10 documentos.

```python
primeros_7_dias = pd.DataFrame(Companies.find({"founded_day" : {"$lte": 7}, }).sort({'acquisition.price_amount':-1}).limit(10))
primeros_7_dias
```

### 16. Todas las empresas en la categoría 'web' que tienen más de 4000 empleados. Ordénalas por la cantidad de empleados en orden ascendente.

```python
web_4000 = pd.DataFrame(Companies.find({'number_of_employees': {"$gt": 4000}}).sort({'number_of_employees':1}))

web_4000
```

### 17. Todas las empresas cuyo monto de adquisición es superior a 10.000.000, y cuya moneda es 'EUR'.

```python
adcqu_EUR = pd.DataFrame(Companies.find({'acquisition.price_amount': {"$gt": 10000000}, 'acquisition.price_currency_code': 'EUR'}))
adcqu_EUR
```

### 18. Todas las empresas que han sido adquiridas en el primer trimestre del año. Limita la búsqueda a 10 empresas, y recupera solo los campos `name` y `acquisition`.

```python
primer_trimestre = pd.DataFrame(Companies.find({'acquisition.acquired_month': {"$lte": 3}}, {'name':1, 'acquisition':1}).limit(10))

# despues_2010 = pd.DataFrame(Companies.find({"acquisition.acquired_year" : {"$gt": 2010}}, {'name':1, 'acquisition': 1}).sort({'acquisition.price_amount':-1}))
primer_trimestre
```
